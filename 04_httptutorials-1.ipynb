```python
import requests
import json

response = requests.get('https://dog.ceo/api/breeds/list/all')
data = response.json()  # Assuming the response is JSON
print(data)
print('-------one record per line----')

# Print each record on a new line in JSON format 

for record in data['message']:
    print(json.dumps(record))
```

```python
# GET Request with Query Parameters
# To send a GET request with query parameters:

import requests
params = {'userId': '1', 'id': '2'}
response = requests.get('url = "https://jsonplaceholder.typicode.com/posts"
', params=params)
data = response.json()
print(data)
# Print each record on a new line in JSON format -Observe the double quotes around keys and values
for record in data:
    print(json.dumps(record))

```

```python
# Handling HTTP Errors
# To handle possible HTTP errors gracefully:

import requests
response = requests.get('https://dog.ceo/api/breeds/list/all')
try:
    response.raise_for_status()  # Raises an HTTPError if the status is 4xx, 5xx
    data = response.json()
    print(data)
except requests.exceptions.HTTPError as err:
    print(f'HTTP Error: {err}')
```

```python
# Setting Timeout for Requests
# To set a timeout for API requests to avoid hanging indefinitely:
import requests
try:
    response = requests.get('https://jsonplaceholder.typicode.com/posts', timeout=5)  # Timeout in seconds
    data = response.json()
    print(data)
except requests.exceptions.Timeout:
    print('The request timed out')

```

```python
# Using Headers in Requests
# To include headers in your request (e.g., for authorization):

import requests
headers = {'Authorization': 'Bearer YOUR_ACCESS_TOKEN'}
response = requests.get('https://jsonplaceholder.typicode.com/posts', headers=headers)
data = response.json()

# Printing headers
print("Headers:")
for key, value in response.headers.items():
    print(f"{key}: {value}") 

# Printing Data
for record in data:
    print(json.dumps(record))
    
```

```python
# POST Request with JSON Payload
# To send data to an API endpoint using a POST request with a JSON payload:
import requests

# Example payload and headers
payload = {'userId': 1, 'title': 'My record' }
headers = {'Content-Type': 'application/json'}

# Sending a POST request
response = requests.post('https://jsonplaceholder.typicode.com/posts', json=payload, headers=headers)

# Checking if the request was successful
if response.status_code == 201:  # 201 Created
    print("Post created successfully:")
    print(response.json())
else:
    print(f"Failed to create post: {response.status_code}")

# Printing Data
params = {'userId': '1'}
get_response = requests.get('https://jsonplaceholder.typicode.com/posts', params=params)


data = get_response.json()
# Printing Data
for record in data:
    print(json.dumps(record))
    
# Ideally you should see the newly created record 9101) in the output.
# due to the nature of the jsonplaceholder.typicode.com API, which is a fake online REST API for testing and prototyping. 
# It simulates POST requests but doesn't actually persist the data. 
# Thus, you won't see the new record when you perform a GET request after a POST request.
'''
In a real-world scenario, you would use an actual backend service where the POST request persists data, and a subsequent GET request would retrieve that data. 
The jsonplaceholder API, however, is only for prototyping and does not maintain state between requests.
'''
```

```python
# Streaming Large Responses
# To stream a large response to process it in chunks, rather than loading it all into memory:

import requests
response = requests.get('https://jsonplaceholder.typicode.com/posts', stream=True)
for chunk in response.iter_content(chunk_size=10):
    # process(chunk)  # Replace 'process' with your actual processing function
    print(chunk)
    
    # To print each chunk in a separate line, you can simply decode each chunk from bytes to a string and print it. 
    if chunk: 
        print(chunk.decode('utf-8'))

# Observe the data with all end of the line characters 
```
